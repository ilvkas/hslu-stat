```python
x = np.array([10, 8, 13, 9, 11, 14, 6, 4, 12, 7, 5])
y1 = np.array([8.04, 6.95, 7.58, 8.81, 8.33, 9.96, 7.24, 4.26, 10.84, 4.82, 5.68])
y2 = np.array([9.14, 8.14, 8.74, 8.77, 9.26, 8.10, 6.13, 3.10, 9.13, 7.26, 4.74])
y3 = np.array([7.46, 6.77, 12.74, 7.11, 7.81, 8.84, 6.08, 5.39, 8.15, 6.42, 5.73])
x4 = np.array([8, 8, 8, 8, 8, 8, 8, 19, 8, 8, 8])
y4 = np.array([6.58, 5.76, 7.71, 8.84, 8.47, 7.04, 5.25, 12.50, 5.56, 7.91, 6.89])
```

### a)

```python
plt.subplot(221)
plt.scatter(x=x, y=y1)
b, a = np.polyfit(x,y1,deg=1)
xl=np.linspace(0,20)
#xl=np.linspace(x.min(),x.max())
plt.plot(xl, a+b*xl, c="orange")

plt.subplot(222)
plt.scatter(x=x, y=y2)
b, a = np.polyfit(x,y2,deg=1)
xl=np.linspace(0,20)
#xl=np.linspace(x.min(),x.max())
plt.plot(xl, a+b*xl, c="orange")

plt.subplot(223)
plt.scatter(x=x, y=y3)
b, a = np.polyfit(x,y3,deg=1)
xl=np.linspace(0,20)
#xl=np.linspace(x.min(),x.max())
plt.plot(xl, a+b*xl, c="orange")

plt.subplot(224)
plt.scatter(x=x4, y=y4)
b, a = np.polyfit(x4,y4,deg=1)
xl=np.linspace(0,20)
#xl=np.linspace(0,x4.max())
plt.plot(xl, a+b*xl, c="orange")

plt.show()
```

- Betrachtet man die vier Streudiagramme (hier noch skaliert), so sieht man, dass nur im ersten Fall eine lineare Regression korrekt ist. <br/>
- Im zweiten Fall ist die Beziehung zwischen X und Y nicht linear, sondern quadratisch. <br/>
- Im dritten Fall gibt es einen Ausreisser, welcher die geschtzten Parameter stark beeinflusst.<br/>
- Im vierten Fall wird die Regressionsgerade durch einen einzigen Punkt bestimmt.

### b)

```python
b1, a1 = np.polyfit(x,y1,deg=1)
b2, a2 = np.polyfit(x,y2,deg=1)
b3, a3 = np.polyfit(x,y3,deg=1)
b4, a4 = np.polyfit(x4,y4,deg=1)
print(b1, a1)
print(b2, a2)
print(b3, a3)
print(b4, a4)
```

Alle Streudiagramme haben praktisch dieselbe Regressionsgerade, obwohl die Streudiagramme sehr unterschiedlich aussehen.<br/>
<b>Fazit:</b> Es gengt nicht, nur bb0 und bb1 anzuschauen. In allenModellen sind diese Schtzungen fast gleich, aber die Datenstze sehen ganz unterschiedlich aus. Eine (graphische) berprfung der Modellannahmen ist also unumgnglich.

### c)

```python
print(np.corrcoef(x,y1))
print(np.corrcoef(x,y2))
print(np.corrcoef(x,y3))
print(np.corrcoef(x4,y4))
```
