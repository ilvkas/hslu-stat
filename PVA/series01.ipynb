### a)

```python
bauer.sum()
```

```python
bauer2 = bauer**2
bauer2.sum()
```

### b)

```python
bauer.size
```

```python
mean = (bauer.sum() / bauer.size)
print(mean)
```

```python
var = (bauer2.sum() / bauer.size) - mean**2
print(var)
```

```python
sigma = np.sqrt(var)
print (sigma)
```
