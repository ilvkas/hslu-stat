# 5.2
### a)

```python
# moegliche Werte von X
werte = np.array([0,10,11])
# X simulieren
sim = pd.Series(np.random.choice(werte, size=1000, replace=True))
# Mehrere Grafiken neben- und untereinander
plt.subplot(4,2,1)
# Histogramm erstellen
sim.hist(bins=[0,1,10,11,12], edgecolor="black", figsize=(10,10))
plt.title("Original")
plt.subplot(4,2,2)
# Normalplot erstellen
st.probplot(sim,plot=plt)
plt.title("Normal Q-Q Plot")
```

### b)

```python
n = 5
# X_1,...,X_n simulieren und in einer n-spaltigen
# Matrix (mit 1000 Zeilen) anordnen
sim = np.random.choice(werte, size=n*1000, replace=True)
sim = pd.DataFrame(np.reshape(sim,(n,1000)))
#In jeder Matrixzeile Mittelwert berechnen
sim_mean = sim.mean()
plt.subplot(4,2,3)
sim_mean.hist(edgecolor="black", figsize=(10,10))
plt.title("Mittelwerte von 5 Beobachtungen")
plt.subplot(4,2,4)
st.probplot(sim_mean,plot=plt)
plt.title("Normal Q-Q Plot")
```

### c)

```python
n = 10
# X_1,...,X_n simulieren und in einer n-spaltigen
# Matrix (mit 1000 Zeilen) anordnen
sim = np.random.choice(werte, size=n*1000, replace=True)
sim = pd.DataFrame(np.reshape(sim,(n,1000)))
#In jeder Matrixzeile Mittelwert berechnen
sim_mean = sim.mean()
plt.subplot(4,2,3)
sim_mean.hist(edgecolor="black", figsize=(10,10))
plt.title("Mittelwerte von 10 Beobachtungen")
plt.subplot(4,2,4)
st.probplot(sim_mean,plot=plt)
plt.title("Normal Q-Q Plot")
```

### b)

```python
iron2 = iron.copy()
```

```python
ironlog = np.log(iron2)
print(ironlog)
```

```python
ironlog.plot(kind="box")
```

Wenn man die Daten logarithmiert, so wird die Varianz stabilisiert, d. h. alle
Gruppen zeigen jetzt eine hnlich grosse Streuung. Der Unterschied in der Lage
ist immer noch ersichtlich.

### c)

```python
plt.subplot(131)
st.probplot(iron["medium"], plot=plt)

plt.subplot(133)
st.probplot(ironlog["medium"], plot=plt)

plt.show()
```

Bei normalverteilten Daten sollte man im Normalplot ungefhr eine Gerade erhalten.
Aus den Plots sieht man, dass dies bei den ursprnglichen Daten (oben)
nicht der Fall ist, whrendman in der Abbildungmit den logarithmierten Daten
(unten) eher eine Gerade erkennt. Die Lognormalverteilung scheint also besser
zu passen.

### c)

```python
zeiten = np.array([16.9, 4.20, 6.70, 8.83, 10.7, 22.4, 1.37, 3.00, 4.82, 4.53, 6.77, 4.81])
```

```python
zs = np.sort(zeiten)
print(zs)
```

```python
zs.size
```

```python
alphak = (np.arange(1, zs.size + 1) - 0.5) / zs.size
print(alphak)
```

```python
quantile_th = st.expon.ppf(q=alphak, scale=1)
print(quantile_th)
```

```python
plt.xlabel("th Q")
plt.ylabel("emp Q")
plt.plot(quantile_th, zs, "o")
x1 = np.linspace(0,3)
y1 = x1*7.4
plt.plot(x1,y1)
plt.show()
```
